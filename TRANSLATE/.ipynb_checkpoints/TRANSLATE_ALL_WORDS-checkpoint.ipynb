```python
import pandas as pd
import nemo
import nemo.collections.nlp as nemo_nlp
```

```python
nmt_model = nemo_nlp.models.machine_translation.MTEncDecModel.from_pretrained(model_name="nmt_es_en_transformer12x2").cuda()
```

```python
all_words_es = pd.read_csv('ALL_WORDS_ES.csv')
```

```python
all_words_es
```

```python
for ind in all_words_es.index:
    translate = nmt_model.translate([all_words_es['WORDS'][ind]])
    all_words_es['TRANSLATE'][ind] = translate[0]
```

```python
all_words_es
```

```python
all_words_es.to_csv('ALL_WORDS_ES.csv')
```
